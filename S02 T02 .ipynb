### Exercici 5
#### Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
import numpy as np
matrix5 = np.array([[2,9,8],[6,1,9],[2,5,3]]) 
fila = 0
for i in matrix5:
    sumCol = 0
    sumFila = 0
    col = 0
    for j in i:
        sumFila += matrix5[fila][col] 
        sumCol += matrix5[col][fila] 
        col += 1
            
    fila += 1
    print("Fila {}: {}".format(fila,sumFila)) 
    print("Columna {}: {}".format(fila,sumCol)) 
    break
```

### Exercici 6
#### Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.

#### Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
res = matrix5%4
print(res)
matrix6 = np.array(res, dtype = bool)
fila = 0
for i in matrix6:
    col = 0
    for j in i:
        if res[fila][col] == 0: matrix6[fila][col] = True
        else: matrix6[fila][col] = False
        col += 1
    fila += 1
print(matrix6)
```

### Exercici 8

#### Mostreu-me a veure què passa quan eliminem el canal G Verd o B Blau. Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.

#### Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
img = mpimg.imread("imag.jfif")
print(img)
pixels = np.array(img)
pixels1 = np.array(img)
for x in range(0,len(pixels)):
    for y in range(0,len(pixels[0])):
        rgb = pixels[x][y]
        rgb[0] = rgb[0] #canal vermell
        rgb[1] = 0 #canal verd
        rgb[2] = rgb[2] #canal blau
        pixels[x][y] = rgb
        
imgplot = plt.imshow(pixels)
mpimg.imsave("imagen1.jpg", pixels)

for x in range(0,len(pixels)):
    for y in range(0,len(pixels[0])):
        rgb = pixels1[x][y]
        rgb[0] = rgb[0] #canal vermell
        rgb[1] = rgb[1] #canal verd
        rgb[2] = 0 #canal blau
        pixels1[x][y] = rgb
        
imgplot = plt.imshow(pixels1)
mpimg.imsave("imagen2.jpg", pixels1)
```
